```python
import tkinter as tk
from tkinter import Button, Label
from PIL import Image, ImageDraw, ImageTk
import numpy as np
from tensorflow.keras.models import load_model
```

```python
#............................................................................................
import tkinter as tk
from tkinter import Button, Label
from PIL import Image, ImageDraw, ImageTk
import numpy as np
import cv2
import pyscreenshot as ImageGrab
from tensorflow.keras.models import load_model

class DrawingApp:
    def __init__(self, root, model):
        self.root = root
        self.model = model
        self.drawn_data = []

        # Create canvas for drawing
        self.canvas = tk.Canvas(root, width=200, height=200, bg="black")
        self.canvas.pack()

        # Create label to display prediction
        self.prediction_label = Label(root, text="Predicted: -", fg="white", bg="black")
        self.prediction_label.pack()

        # Create buttons for actions
        self.button_predict = Button(root, text="Predict", command=self.handle_button_click_predict)
        self.button_erase = Button(root, text="Erase", command=self.handle_button_click_erase)
        self.button_display = Button(root, text="Display Image", command=self.handle_button_click_display)

        self.button_predict.pack(side="left", padx=10)
        self.button_erase.pack(side="left", padx=10)
        self.button_display.pack(side="left", padx=10)

#         # Initialize PhotoImage attribute
#         self.drawn_image_photo = None

        # Binding mouse events
        self.canvas.bind("<Button-1>", self.start_draw)
        self.canvas.bind("<B1-Motion>", self.draw)

    def start_draw(self, event):
        self.drawn_data = [(event.x, event.y)]

    def draw(self, event):
        self.drawn_data.append((event.x, event.y))
        self.canvas.create_line(self.drawn_data[-2], self.drawn_data[-1], fill="white", width=5)

    def predict_character(self):
        # Capture the image using ImageGrab
        img = ImageGrab.grab(bbox=(22, 29, 220, 226))

        # Convert the Image to a NumPy array
        img_np = np.array(img)

        # Convert the image to grayscale using OpenCV
        img_gray = cv2.cvtColor(img_np, cv2.COLOR_BGR2GRAY)

        # Resize the image to (64, 64)
        img_resized = cv2.resize(img_gray, (64, 64))

        # Normalize pixel values to [0, 1]
        img_normalized = img_resized / 255.0

        # Add batch and channel dimensions
        img_array = np.expand_dims(img_normalized, axis=0)
        img_array = np.expand_dims(img_array, axis=-1)

        # Make a prediction
        try:
            prediction = self.model.predict(img_array)
            predicted_class = np.argmax(prediction)
            probability = np.max(prediction) * 100  # Probability in percentage
            predicted_character = index_to_char[predicted_class]
            return predicted_character, probability
        except Exception as e:
            print(f"Prediction error: {e}")
            return "Error", 0

    def display_drawn_image(self):
        img = ImageGrab.grab(bbox=(22,29,220,226))

        img.save("paint.jpg")

        im = cv2.imread("paint.jpg")
        load = Image.open("paint.jpg")
        load= load.resize((280,280))
        photo = ImageTk.PhotoImage(load)
        
        # Display the drawn image in a separate window
        drawn_image_window = tk.Toplevel(self.root)
        drawn_image_label = tk.Label(drawn_image_window, image=photo)
        drawn_image_label.image = photo
        drawn_image_label.pack()

    def handle_button_click_predict(self):
        predicted_character, probability = self.predict_character()
        self.prediction_label.config(text=f"Predicted: {predicted_character}, Probability: {probability:.2f}%")

    def handle_button_click_erase(self):
        self.canvas.delete("all")
        self.prediction_label.config(text="Predicted: -")

    def handle_button_click_display(self):
        self.display_drawn_image()

# Load your pre-trained model
model = load_model('nepali_model.h5')

# Assuming you have a dictionary of class indices to character labels
index_to_char = {
    0: 'क', 1: 'ख', 2: 'ग', 3: 'घ', 4: 'ङ', 5: 'च', 6: 'छ', 7: 'ज', 8: 'झ', 9: 'ञ', 
    10: 'ट', 11: 'ठ', 12: 'ड', 13: 'ढ', 14: 'ण', 15: 'त', 16: 'थ', 17: 'द', 18: 'ध',
    19: 'न', 20: 'प', 21: 'फ', 22: 'ब', 23: 'भ', 24: 'म', 25: 'य', 26: 'र', 27: 'ल',
    28: 'व', 29: 'श', 30: 'ष', 31: 'स', 32: 'ह', 33: 'क्ष', 34: 'त्र', 35: 'ज्ञ',
    36: '०', 37: '१', 38: '२', 39: '३', 40: '४', 41: '५', 42: '६', 43: '७', 44: '८', 45: '९'
}


# Create tkinter window
root = tk.Tk()
root.title("Draw Devanagari Characters")

# Create the DrawingApp instance
app = DrawingApp(root, model)

root.mainloop()
```
